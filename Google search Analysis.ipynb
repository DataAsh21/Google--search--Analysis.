## Google Search Analysis with Python

On average, 3.5 billion searches are made every day on Google, which translates to about 40,000 queries made every second. Therefore, using Google search to analyse data based on search queries is an excellent use case.

Google doesn't provide much access to information regarding daily search queries, but Google Trends, another Google service, can be utilised to perform Google search analysis. An API is available from Google Trends that may be used to examine the daily Google searches. This API is called pytrends, and using the pip command, pip install pytrends, you can quickly install it on your computer.

Now that we are successfully installed the pytrends library on your computer, let's begin the work of analysing Google search results by importing the required Python libraries:

```python
import pandas as pd
from pytrends.request import TrendReq
import matplotlib.pyplot as plt
import seaborn as sns
pytrends = TrendReq()
```

Here I will be analyzing the Google search trends on the queries based on “Data Analyst”, "Data Scientist","Machine Learning" so let’s create a DataFrame of the top 10 countries which search for “Data Analyst” on Google:

#### Search Interest Over Time: 

```python
# Set up search parameters
keywords = ['Data Analyst', 'Data Scientist','Machine Learning']
pytrends.build_payload(kw_list=keywords, timeframe='today 5-y', geo='IN')

# Get Google Trends data for each keyword
data = pytrends.interest_by_region()
data = data.sort_values(by=keywords, ascending=False)
data = data.head(10)

print(data)
```

*The analysis of search interest over time revealed the following trends:*

*The search interest for "Data Analyst" has shown a steady increase over the past five years in India. This indicates a growing interest in data analytics roles among users.*

*The search interest for "Data Scientist" has exhibited a significant surge in the past two years, indicating a rising demand for data science professionals in India.*

*The search interest for "Machine Learning" has witnessed a remarkable upward trend, demonstrating the increasing curiosity and relevance of this field in India.*

#### Regional Variation:

```python
# Plot the data
data.plot(kind='bar', figsize=(10, 8))
plt.title('Google Search Interest by Region')
plt.xlabel('Region')
plt.ylabel('Search Interest')
plt.xticks(rotation=45)
plt.legend(keywords, loc='upper left')
plt.show()
```

*Analyzing the regional interest in these keywords highlighted the following insights:*

* *The states of Maharashtra, Kerala, and Delhi show the highest search interest for all three keywords, indicating significant opportunities for digital marketing companies in these regions.*

* *Other regions, such as Tamil Nadu, Telangana, and Uttar Pradesh, also demonstrate considerable interest, suggesting potential markets for targeted marketing campaigns and services.*

* *It is essential for digital marketing companies to tailor their strategies based on regional preferences and target areas with higher search interest.*

#### Keyword Comparison

```python
# Get Google Trends data
trends_data = pytrends.interest_over_time()
# Plot the trends
plt.figure(figsize=(65, 20))
trends_data.plot()
plt.title('Google Search Trends')
plt.xlabel('Date')
plt.ylabel('Search Interest')
plt.legend(keywords)
plt.show()
```

*Comparing the search interest for different keywords reveals valuable insights:*

* *The search interest for "data scientist" has surpassed that of "data analyst" and "machine learning" in recent years, indicating a shift in user preferences and market demand.*

* *However, "data analyst" continues to maintain a stable and consistent level of interest, signifying a sustained need for professionals in this field.*

* *"Machine learning" exhibits the fastest growth rate among the three keywords, suggesting an increasing curiosity and adoption of this technology.*
